```python
def age_group (x):
    if x['patient_age'] <10  :
        return '0-10'
    if x['patient_age'] >= 10 & x['patient_age'] <20   :
        return '10-19'
    if x['patient_age'] >=20 & x['patient_age'] <30    :
        return '21-29'
    if x['patient_age'] >= 30 & x['patient_age'] <40   :
        return '31-39'
    if x['patient_age'] >= 40 & x['patient_age'] <50   :
        return '41-49'
    if x['patient_age'] >= 50 & x['patient_age'] <60   :
        return '51-59'
    if x['patient_age'] >= 60 & x['patient_age'] <70   :
        return '61-69'
    if x['patient_age'] >= 70 & x['patient_age'] <80   :
        return '71-79'
    if x['patient_age'] >= 80 & x['patient_age'] <90   :
        return '91-89'
    if x['patient_age'] >= 90                        :   
        return '90+' 
    else                                           :
        return 'No'
```
